# Linear Regression

### Modell

* Pro Datenpunkt/Beispiel: $\hat{y}_i = \sum_{j=1}^p \beta_j x_{i,j} = x_i \cdot \beta$
* In Matrixschreibweise: $Y = X \cdot \beta$

```python
# a (n,p) matrix of number randomly drawn from the interval (0,1)
n = 21 # number of rows == number of examples == number of observations == number of data points
p = 2  # number of columns == number of variables == number of features == number of attributes
X = np.random.rand(n, p)
X
```

```python
# lets predend the data we get is generated by linear model with this weight vector
beta_true = np.array([-1.3, +1.5])
beta_true
```

```python
# so the true/observed Y values would look like the following
y_true = X @ beta_true
y_true
```

#### Detour: A visualization using matplotlib

```python
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

# Make data.
x0 = np.arange(-0.25, 1.25, 0.1)
x1 = np.arange(-0.25, 1.25, 0.1)

X0, X1 = np.meshgrid(x0, x1)
YG = X0 * beta_true[0] + X1*beta_true[1]

ax.view_init(30, 30)
# Plot the model
ax.plot_surface(X0, X1, YG, cmap=cm.gray, linewidth=1, alpha=1, antialiased=False)

# Plot the data points
ax.plot(X[:,0], X[:,1], y_true, 'r.');
```

### Learning a liner model using ordinary least squares

* Find $\beta^*$ such that the sum of squared errors is minimal

$$
  \beta^* = argmin_\beta  \sum_{i=1}^n ( \beta \cdot x_i - y_i)^2 = argmin_\beta (X \cdot \beta - Y)^T \cdot (X \cdot \beta - Y)
$$

* Solution (will be derived at the blackboard):

$$
\beta^* = (X^T \cdot X)^{-1} \cdot X^T \cdot Y
$$

```python
beta_star = np.linalg.inv(X.T @ X) @ (X.T @ y_true)
beta_star
```

### Definieren von Funktionen

```python
def learn_linear_regression(X, y):
    return np.linalg.inv(X.T @ X) @ (X.T @ y)

def apply_linear_regression(X, beta):
    return np.dot(X, beta)
```

```python
beta_star = learn_linear_regression(X, y_true)
y_pred = apply_linear_regression(X,beta_star)
```

### Berechnen des MSE und RMSE

```python
# calculate the MSE
mean_squared_error = np.mean((y_pred - y_true)**2)

# calculate the RMSE
root_mean_squared_error = np.sqrt(mean_squared_error)
root_mean_squared_error
```

## Now we make it more realistic: There is always noise in the data

```python
# generate obervations based on beta_true but add some noise
y_0 = X @ beta_true

# noise is drawn from a  Gaussian distribution with a standard deviation of sigma
sigma = 0.11
noise = sigma * np.random.randn(n)

# add the noise
y_true = y_0 + noise
```

```python
x=np.array(range(n))
plt.plot(x, y_0, x, noise, x, y_true);
plt.grid(True)
```

```python
beta_star = learn_linear_regression(X, y_true)
y_pred = apply_linear_regression(X,beta_star)

mean_squared_error = np.mean((y_pred - y_true)**2)
root_mean_squared_error = np.sqrt(mean_squared_error)
root_mean_squared_error
```
